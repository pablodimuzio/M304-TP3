```python
def suite(g, epsilon):
    u = 1
    nombre_passages = 1
    while np.abs(u-g(u))>=epsilon:
        u = g(u)
        nombre_passages += 1
    return nombre_passages, N(u), N(g(u))
```

On nomme la variable interne `g` pour éviter toute confusion avec la fonction prédéfinit `f`.

```python
def dicho(g, a, b, eps):
    
    nombre_passages = 0
    
    while np.abs( g(a) - g(b) ) >= eps:
        
        nouvelle_borne = (a+b)/2
        
        a = nouvelle_borne if np.sign(g(nouvelle_borne)) == np.sign(g(a)) else a
        b = nouvelle_borne if np.sign(g(nouvelle_borne)) == np.sign(g(b)) else b
        
        nombre_passages += 1
    return a,b, nombre_passages
```

```python
dicho(lambda x : f(x)-x, 1, 2, 10**-10)
```

On obtient une approximation à $10^{-10}$ près après 36 itérations soit 10 de plus que pour l'autre méthode, on la préférera donc pour cet exemple.

```python
def serie(eps):
    n = 1
    S = 0
    erreur = 1/(4*(n+1)**2-1)
    while n < 10**6 and erreur >= eps:
        
        S += ((-1)**(n+1))/(4*n**2-1)
        n += 1
        
        erreur = 1/(4*(n+1)**2-1)
        
    return S, N(erreur), n
```
